```python
import pandas as pd
import numpy as np

excel_measurements = pd.read_excel("NSF-IOS_2017Measurements_kcb.xlsx")

# imaged were downsized from 6000 as a max dimension to 1200 as a max dimension so pixels are 5x as large in meters
pixel_size_factor = 6000 / 1200

measurements_df = pd.DataFrame(np.nan, index=np.arange(0, len(excel_measurements)), columns=['Filename', 'Auto_Measurment', 'Manual_Measurement', 'Estimated_Manual_Error'])

measurements_df.set_index("Filename")

for index, row in excel_measurements.iterrows():
    # TODO change this to be your count make sure it matches up with the image filenames match up
    cnn_pixel_count = 500
    
    total_altitude = row["Total Altitude"]
    pixel_size = row["Pixel size"]
    focal_length = row["Focal length (mm)"]
    
    adjusted_cnn_pixel_count = pixel_size_factor * cnn_pixel_count
        
    measurements_df.loc[measurements_df.index[index], "Filename"] = row["Image"]
    measurements_df.loc[measurements_df.index[index], "Auto_Measurment"] = adjusted_cnn_pixel_count * pixel_size/focal_length * total_altitude
    measurements_df.loc[measurements_df.index[index], "Manual_Measurement"] = row["Total Length (m)"]
    measurements_df.loc[measurements_df.index[index], "Estimated_Manual_Error"] = row["Estimated % Error"]
    
measurements_df
```
