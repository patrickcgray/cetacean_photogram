```python
import pandas as pd
import numpy as np

excel_measurements = pd.read_excel("NSF-IOS_2017Measurements_kcb.xlsx")

# imaged were downsized from 6000 as a max dimension to 1200 as a max dimension so pixels are 5x as large in meters
pixel_size_factor = 6000 / 1200

measurements_df = pd.DataFrame(np.nan, index=np.arange(0, len(excel_measurements)), columns=['Filename', 'Auto_Measurment', 'Manual_Measurement', 'Estimated_Manual_Error'])

measurements_df.set_index("Filename")

for index, row in excel_measurements.iterrows():
    # TODO change this to be your count make sure it matches up with the image filenames match up
    cnn_pixel_count = 500
    
    total_altitude = row["Total Altitude"]
    pixel_size = row["Pixel size"]
    focal_length = row["Focal length (mm)"]
    
    adjusted_cnn_pixel_count = pixel_size_factor * cnn_pixel_count
        
    measurements_df.loc[measurements_df.index[index], "Filename"] = row["Image"]
    measurements_df.loc[measurements_df.index[index], "Auto_Measurment"] = adjusted_cnn_pixel_count * pixel_size/focal_length * total_altitude
    measurements_df.loc[measurements_df.index[index], "Manual_Measurement"] = row["Total Length (m)"]
    measurements_df.loc[measurements_df.index[index], "Estimated_Manual_Error"] = row["Estimated % Error"]
    
measurements_df
```

```python
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np

ia.seed(1)

# Example batch of images.
# The array has shape (32, 64, 64, 3) and dtype uint8.
images = np.array(
    [ia.quokka(size=(64, 64)) for _ in range(32)],
    dtype=np.uint8
)

seq = iaa.Sometimes(.667, iaa.Sequential([
    iaa.Fliplr(0.5), # horizontal flips
    iaa.Crop(percent=(0, 0.1)), # random crops
    # Small gaussian blur with random sigma between 0 and 0.25.
    # But we only blur about 50% of all images.
    iaa.Sometimes(0.5,
        iaa.GaussianBlur(sigma=(0, 0.25))
    ),
    # Strengthen or weaken the contrast in each image.
    iaa.ContrastNormalization((0.75, 1.5)),
    # Add gaussian noise.
    # For 50% of all images, we sample the noise once per pixel.
    # For the other 50% of all images, we sample the noise per pixel AND
    # channel. This can change the color (not only brightness) of the
    # pixels.
    iaa.AdditiveGaussianNoise(loc=0, scale=(0.0, 0.05*255)),
    # Make some images brighter and some darker.
    # In 20% of all cases, we sample the multiplier once per channel,
    # which can end up changing the color of the images.
    iaa.Multiply((0.8, 1.2)),
    # Apply affine transformations to each image.
    # Scale/zoom them, translate/move them, rotate them and shear them.
    iaa.Affine(
        scale={"x": (0.8, 1.2), "y": (0.8, 1.2)},
        #translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)},
        rotate=(-180, 180),
        #shear=(-8, 8)
    )
], random_order=True)) # apply augmenters in random order

images_aug = seq.augment_images(images)
```

```python
augmentation = iaa.Sometimes(0.9, [
        iaa.Fliplr(0.5),
        iaa.Flipud(0.5),
        iaa.Multiply((0.8, 1.2)),
        iaa.GaussianBlur(sigma=(0.0, 5.0))
        ])
```
